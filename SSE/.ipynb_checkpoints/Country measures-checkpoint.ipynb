```python
import pandas as pd
import numpy as np
```

```python
# Importo primero los países, y voy a ir agregando columnas a partir de los diferentes datasets
paises=pd.read_csv('data/countries.csv')
paises
```

```python
# Importo las medidas
# Este dataset está bajado de 
# https://www.acaps.org/covid19-government-measures-dataset
# Y tiene una fila por cada medida tomada en cada país
df=pd.read_csv('data/measures.csv', parse_dates=['DATE_IMPLEMENTED'])
display(df.head())


```

```python
# ESto lo usé solamente una vez para generar las parejas país, ISO
#df.groupby(['COUNTRY', 'ISO']).first().index.to_frame().reset_index(drop=True).to_csv('data/countries.csv', index=False)
```

```python
# Me quedo solamented con el comienzo de las medidas, y pongo en cada columna la medida
# En la celda queda la fecha de la medida, si es que se tomó
# Además elijo solamente algunas columnas relevantes
dataset=df.loc[df['LOG_TYPE']!='Phase-out measure',['ISO','MEASURE','DATE_IMPLEMENTED']]
medidas=dataset.pivot_table(index='ISO', columns='MEASURE', values='DATE_IMPLEMENTED', aggfunc=np.min)
medidas.reset_index(inplace=True)

# Le agrego el nombre del país, a partir de la tabla de ISO
medidas=pd.merge(medidas, paises, on='ISO', how='left')
medidas
```

```python
# Importo la fecha de primer contagiado
contagiados=pd.read_csv('data/first_cases.csv')
contagiados
```
